# Optimizer comparison on a regression task using Diabetes dataset and Ridge Regression

```python
import torch
import torch.nn as nn

from src.quadratic_annealing_optimizer import QuadraticAnnealingOptimizer
from src.models import Ridge
from src.losses import RidgeLoss
from src.utils import data_load_and_prep, build_sampler
from src.training import train
from src.newton_optimizer import NewtonOptimizer

EXPERIMENT_NAME = "optimizer-comparison-ridge-diabetes"
```

```python
train_loader, test_loader = data_load_and_prep("diabetes", test_size=0.3, random_state=42, batch_size='full')
```

## Adam

```python
loss_fn = RidgeLoss(alpha=1e2) 
adam_model = Ridge(input_dim=10, output_dim=1)
classical_device = "cuda" 
adam_optimizer = torch.optim.Adam(adam_model.parameters(), 
                             lr=3,
                             betas=[0.9, 0.999],
                             )
```

## Adam optimizaer

```python
train(
    model=adam_model,
    train_loader=train_loader,
    test_loader=test_loader,
    loss_fn=loss_fn,
    optimizer=adam_optimizer,
    c_device=classical_device,
    epochs=100, 
    experiment_name = EXPERIMENT_NAME,
    run_name="adam-optimizer"
)
```

## l-BFGS optimizer

```python
loss_fn = RidgeLoss(alpha=1e2)
lbfgs_model = Ridge(input_dim=10, output_dim=1)
classical_device = "cuda" 
lbfgs_optimizer = torch.optim.LBFGS(lbfgs_model.parameters(), 
                              lr=0.5,
                              max_iter=1,
                              history_size=100,
                              line_search_fn=None,
                              )
```

```python
train(
    model=lbfgs_model,
    train_loader=train_loader,
    test_loader=test_loader,
    loss_fn=loss_fn,
    optimizer=lbfgs_optimizer,
    c_device=classical_device,
    epochs=100,
    experiment_name = EXPERIMENT_NAME,
    run_name="lbfgs-optimizer"
)
```

## Newton-Raphson optimizer

```python
loss_fn = RidgeLoss(alpha=1e2)
newton_model = Ridge(input_dim=10, output_dim=1)
classical_device = "cuda" 
newton_optimizer = NewtonOptimizer(newton_model.parameters(), 
                              lr=0.1,
                              max_iter=1,
                              damping=1e-4,
                              )
```

```python
train(
    model=newton_model,
    train_loader=train_loader,
    test_loader=test_loader,
    loss_fn=loss_fn,
    optimizer=newton_optimizer,
    c_device=classical_device,
    epochs=100,
    experiment_name = EXPERIMENT_NAME,
    run_name="newton-optimizer"
)
```

## Quadratic Annealing optimizer (cpu)

```python
loss_fn = RidgeLoss(alpha=1e2)
model = Ridge(input_dim=10, output_dim=1)
classical_device = "cpu"

sampler = build_sampler(mode="simulated")

optimizer = QuadraticAnnealingOptimizer(
    sampler=sampler,
    model=model,
    device="cpu",
    subset_size=10,
    step_size=0.75,
    num_reads=100,
)
```

```python
train(
    model=model,
    train_loader=train_loader,
    test_loader=test_loader,
    loss_fn=loss_fn,
    optimizer=optimizer,
    c_device=classical_device,
    epochs=100,
    experiment_name = EXPERIMENT_NAME,
    run_name="quadratic-annealer-cpu-optimizer"
)
```
